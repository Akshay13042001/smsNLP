```python
df2=pd.read_csv('/content/drive/MyDrive/Datasets/SMS_test.csv',encoding='ISO-8859-1')
df2
```

```python
df=pd.concat([df1,df2],axis=0,ignore_index=True)
df
```

```python
df3=df.iloc[:,1:]
df3
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df3['Label']=le.fit_transform(df['Label'])
df3
```

```python
msg=df.Message_body
msg
```

```python
msg=msg.str.replace('[^a-zA-Z0-9]+'," ")
msg
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
stemmer=SnowballStemmer('english')
tk=TweetTokenizer()
```

```python
msg=msg.apply(lambda line:[stemmer.stem(token.lower()) for token in tk.tokenize(line)]).apply(lambda tokens:' '.join(tokens))
msg
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
sw=stopwords.words('english')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
x=vectorizer.fit_transform(msg)
print(x)
```

```python
x.shape
```

```python
x=x.toarray()
x
```

```python
y=df3.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=1)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
rc=RandomForestClassifier()
knn=KNeighborsClassifier(n_neighbors=7,weights='uniform',algorithm='auto')
sv=SVC()
models=[rc,knn,sv]
for model in models:
  print('*******',model,'********')
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  print(classification_report(y_test,y_pred))
  nw=vectorizer.transform(['You’ve won a prize! Go to [link] to claim your $500 Amazon gift card'])
  nw=nw.toarray()
  y=model.predict(nw)
  if y==0:
    print('Not Spam')
  else:
    print('Spam')
```
